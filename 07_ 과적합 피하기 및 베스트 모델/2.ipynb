```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import numpy
import pandas as pd
import tensorflow as tf

# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('../dataset/sonar.csv', header=None)

dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

# 10개의 파일로 쪼갬
#10-> 5로 변경
n_fold = 5
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)

# 빈 accuracy 배열
accuracy = []

i = 0
# 모델의 설정, 컴파일, 실행
for train, test in skf.split(X, Y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs=100, batch_size=5)
    
    save_name = './model/lab7-model-' + str(i) + '.hdf5'
    model.save(save_name)
    #model save
    
    k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])
    accuracy.append(k_accuracy)
    
    i += 1

# 결과 출력
print("\n %.f fold accuracy:" % n_fold, accuracy)

#가장 높은 accuracy를 가진 딥러닝 모델의 index 값을 찾음
index = accuracy.index(max(accuracy))
print('Max index: ' + str(index))
```

```python
from keras.models import load_model

del model
load_name = './model/lab7-model-' + str(index) + '.hdf5'
model = load_model(load_name)

print('Load '+ str(index) + '-th model having ' + str(max(accuracy)) + ' accuracy')
print('Simple test of the first ' + str(n_fold) + ' element')

x_train = dataset[:5,0:60]

y_train = dataset[:5, 60]
e.fit(y_train)
Y_train = e.transform(y_train)

output = model.predict(x_train)

for i in range(5):
    print("index: %d, actual: %.02f, expect: %.02f" % (i, Y[i], output[i]))
```
