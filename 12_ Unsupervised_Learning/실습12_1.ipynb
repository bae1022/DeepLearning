```python
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd

for i in range(4):
    num = i
    
    if (i == 3):
        num = 5 #'IT 과학'은 105번
        
    url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=10" + str(num)
    print(url)
    #url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=101#&date= 00:00:00&page=2" 

    #크롤링 자동 방지 해제 
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
           AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"}
    result = requests.get(url, headers=headers)
    #print(result)

    
    soup = BeautifulSoup(result.text, 'html.parser')
    #print(soup)

    news_titles = soup.find_all(class_='cluster_text_headline')
    #news_titles = soup.select('.cluster_text_headline')
    #print(news_titles)

    print('총 뉴스 제목 개수: ', len(news_titles))
    for item in news_titles:
        print(item.get_text())
        #print(item['href'])
        
    if os.path.isfile('./test_SectionDataset_11.csv'):
        df = pd.read_csv('./test_SectionDataset_11.csv', names=['title', 'section'], header=0)
    
    else:
        bodyStruc = {'title': [], 'section': []}
        df = pd.DataFrame( bodyStruc, columns=['title', 'section'] )
    
    for item in news_titles:
        texts = item.get_text()
        texts = texts.replace(',', "")
        texts = texts.replace('\'', "")
        texts = texts.replace('\"', "")
        texts = texts.replace('…', " ")
        print(texts)
        oneData = {'title': texts, 'section': str(i)}
        df = df.append( oneData, ignore_index=True )

    df.to_csv('./test_SectionDataset_11.csv', encoding='UTF-8', index=False, header=True)
```
