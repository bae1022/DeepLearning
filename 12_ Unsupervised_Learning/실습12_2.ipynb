```python
import os
import pandas as pd

data = pd.read_csv('../output/Article_정치_202012_202012.csv', encoding='euc-kr', delimiter=",")

data_2500 = data.head(2500)

data_2500_title = data_2500.iloc[:, 3].tolist() #기사 제목


if os.path.isfile('./train_SectionDataset_2.csv'):
    df = pd.read_csv('./train_SectionDataset_2.csv', names=['title', 'section'], header=0)
else:
    bodyStruc = {'title': [], 'section': []}
    df = pd.DataFrame( bodyStruc, columns=['title', 'section'] )
    
for i in range(len(data_2500_title)):
    texts = data_2500_title[i]
    texts = texts.replace(',', "")
    texts = texts.replace('\'', "")
    texts = texts.replace('\"', "")
    texts = texts.replace('…', " ")
    print(texts)
    oneData = {'title': texts, 'section': '0'}
    df = df.append( oneData, ignore_index=True )

df.to_csv('./train_SectionDataset_2.csv', encoding='UTF-8', index=False, header=True)
```

```python
import os
import pandas as pd

for i in range(4):

    if (i == 0):
        data = pd.read_csv('../output/Article_정치_202012_202012.csv', header=None, encoding='euc-kr', delimiter=",")
        
    elif (i == 1):
        data = pd.read_csv('../output/Article_경제_202012_202012.csv', header=None, encoding='euc-kr', delimiter=",")
        
    elif (i == 2):
        data = pd.read_csv('../output/Article_사회_202012_202012.csv', header=None, encoding='euc-kr', delimiter=",")
        
    elif (i == 3):
        data = pd.read_csv('../output/Article_IT과학_202012_202012.csv', header=None, encoding='euc-kr', delimiter=",")    
    
    data_2500 = data.head(2500)

    data_2500_title = data_2500.iloc[:, 3].tolist() #기사 제목
    data_2500_section = data_2500.iloc[:, 1].tolist() #기사 영역
    
    if (data_2500_section[i] == '정치'):
        num = 0
    
    elif (data_2500_section[i] == '경제'):
        num = 1
    
    elif (data_2500_section[i] == '사회'):
        num = 2
    
    elif (data_2500_section[i] == 'IT과학'):
        num = 3
    
    if os.path.isfile('./train_SectionDataset_11.csv'):
        df = pd.read_csv('./train_SectionDataset_11.csv', names=['title', 'section'], header=0)
    else:
        bodyStruc = {'title': [], 'section': []}
        df = pd.DataFrame( bodyStruc, columns=['title', 'section'] )
    
    for i in range(len(data_2500_title)):
        texts = data_2500_title[i]
        texts = texts.replace(',', "")
        texts = texts.replace('\'', "")
        texts = texts.replace('\"', "")
        texts = texts.replace('…', " ")
        print(texts)
        oneData = {'title': texts, 'section': str(num)}
        df = df.append( oneData, ignore_index=True )

    df.to_csv('./train_SectionDataset_11.csv', encoding='UTF-8', index=False, header=True)
```
