```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mpl_toolkits import mplot3d

df = pd.read_csv('../dataset/pima-indians-diabetes.csv',
               names = ["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])

tmp = df[['pregnant', 'plasma', 'class']].values.tolist()
data1 = np.array(tmp)
data2 = data1[:5, :]

print(data2)
print(data2.shape)

x = []
y = []
z = []

for i in range(data2.shape[0]):
    x.append(data2[i][0])
    
for i in range(data2.shape[0]):
    y.append(data2[i][1])
    
for i in range(data2.shape[0]):
    z.append(data2[i][2])
    
print(x)
print(y)
print(z)

ax = plt.axes(projection='3d')
ax.plot(x, y, z, "o", c="blue", markersize=6)
ax.plot(x, y, z)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mpl_toolkits import mplot3d

df = pd.read_csv('../dataset/pima-indians-diabetes.csv',
               names = ["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])

tmp = df[['age']].values.tolist()
data1 = np.array(tmp)

sum = 0
for i in range(len(data1)):
    n = data1[i]
    sum += n
    
avg = sum / len(data1)

print("Average age: %0.2f" % (avg))

low_age = df[df['age'] < 33]
high_age = df[df['age'] > 33]

low_age.thickness.plot.hist(bins=100, alpha=1.0)
high_age.thickness.plot.hist(bins=100, alpha=0.4)

plt.legend(['low_age', 'high_age'], loc="upper right")
plt.xlabel('thickness')
plt.show()
```
