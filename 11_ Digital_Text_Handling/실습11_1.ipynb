```python
# -*- coding: utf-8 -*-
# 코드 내부에 한글을 사용가능 하게 해주는 부분입니다.

#import numpy
import tensorflow as tf
from numpy import array
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Embedding
import numpy as np

import pandas as pd
from keras.utils import np_utils

#주어진 문장을 '단어'로 토큰화 하기

#케라스의 텍스트 전처리와 관련한 함수중 text_to_word_sequence 함수를 불러 옵니다.
from tensorflow.keras.preprocessing.text import text_to_word_sequence

data = pd.read_csv('../dataset/train_mydataset_6000.csv', delimiter=",")
docs = data.title
classes = np_utils.to_categorical(data.label)

test = pd.read_csv('../dataset/test_mydataset_1500.csv', delimiter=",")
test_docs = test.title
test_classes = np_utils.to_categorical(test.label)

doc_append = docs.append(test_docs) #X_train + X_test 수행

# 토큰화_train
token = Tokenizer()
#token.fit_on_texts(docs)
token.fit_on_texts(doc_append)
x = token.texts_to_sequences(docs)

# 패딩_train
padded_x = pad_sequences(x, 14)

#임베딩에 입력될 단어의 수를 지정합니다.
word_size = len(token.word_index) +1


print('첫번째 입력: ' + docs[0])
print('첫번째 one-hot 출력: ' + str(classes[0]))
print('첫번째 토큰 결과: ' + str(x[0]))

max_title = -1
for i in range(len(x)):
    if max_title < len(x[i]):
        max_title = len(x[i])
       
print('학습셋 제목 최대 길이: ' + str(max_title))



# 토큰화 _test
#token = Tokenizer()
#token.fit_on_texts(test_docs)
test_x = token.texts_to_sequences(test_docs)

# 패딩_test
test_padded_x = pad_sequences(test_x, max_title)  

max_test_title = -1
for j in range(len(test_x)):
    if max_test_title < len(test_x[j]):
        max_test_title = len(test_x[j])
               
print('테스트셋 제목 최대 길이: ' + str(max_test_title))
print('첫번째 패딩 토큰: ' + str(padded_x[0]))
print('Word Size: ' + str(word_size))

#단어 임베딩을 포함하여 딥러닝 모델을 만들고 결과를 출력합니다.
model = Sequential()
model.add(Embedding(word_size, 8, input_length=max_title))
model.add(Flatten())
model.add(Dense(3, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(padded_x, classes, epochs=20)
```

```python
print("\n Accuracy: %.4f" % (model.evaluate(test_padded_x, test_classes)[1]))
```
