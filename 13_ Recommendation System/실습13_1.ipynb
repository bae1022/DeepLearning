```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv('../dataset/ClusterInput.csv',
                names=["x", "y"])
print( df.describe() )
df2 = df[ ['x', 'y'] ] #두개의 칼럼만 사용

model = KMeans( n_clusters=4, random_state=0 ) #4개의 군집으로 분류
model.fit(df2)

result = model.predict(df2)
print(result)
print("개수: ", len(result) )
```

```python
centers = model.cluster_centers_
print(centers)

plt.scatter( x=df2['x'], y=df2['y'], c=result )
plt.scatter( x=centers[:,0], y=centers[:,1], marker='D', c='r')
```
